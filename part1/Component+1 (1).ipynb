```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
```

```python
with open("/home/hgao13/work/data-readonly/CUMTD/stops.txt", "r") as f:
    reader = csv.reader(f)
    header = next(reader)
    data = {}
    stop = {}
    for column in header:
        stop[column] = []
    for row in reader:
        for column, value in zip(header, row):
            stop[column].append(value)
```

```python
stop
```

```python
plt.plot(stop['stop_lon'], stop['stop_lat'], '.')
```

```python
lat, lon=40.1164, -88.2434
dlat = 0.1
dlon = 0.1
```

```python
import cartopy.crs as ccrs
from cartopy.io.img_tiles import OSM
```

```python
osm_tiles = OSM()

plt.figure(figsize=(12, 12))

# Use the tile's projection for the underlying map.
ax = plt.axes(projection=osm_tiles.crs)

# Specify a region of interest, in this case, CU Area.
ax.set_extent([lon - dlon, lon + dlon, lat - dlat, lat + dlat],
              ccrs.PlateCarree())

ax.add_image(osm_tiles,12)

ax.scatter(list(stop['stop_lon']),list(stop['stop_lat']), transform=ccrs.PlateCarree())
```

The data was loaded and stored as a dictionary. Then, each column was read and plot by their position, which generated the first graph. And finally, it was overlapped onto a map by OSM. According ot this visualization, it is easy to identify the area was covered by CUMTD, as well as the density of the stops. However, the center part has data that overlapped on each other, which is hard to visualize. For further improvements, the buslines should be separated out into different colors to show the different routes; and a more detailed map with main streets names could be better.
